```python
# Analisar período que as pessoas ficaram com a assinatura criando uma nova coluna para averiguar
df_netflix['duracao_assinatura'] = df_netflix['Last Payment Date'] - df_netflix['Join Date']

df_netflix
```

```python
# o .loc faz um filtro especificando a coluna que vamos usar, nesse caso o Age

df_netflix.loc[df_netflix['Age'] < 25, 'grupo_idade'] = '20-24'
df_netflix.loc[(df_netflix['Age'] >= 25) & (df_netflix['Age'] < 30), 'grupo_idade'] = '25-29'
df_netflix.loc[(df_netflix['Age'] >= 30) & (df_netflix['Age'] < 35), 'grupo_idade'] = '30-34'
df_netflix.loc[(df_netflix['Age'] >= 35) & (df_netflix['Age'] < 40), 'grupo_idade'] = '35-39'
df_netflix.loc[(df_netflix['Age'] >= 40) & (df_netflix['Age'] < 45), 'grupo_idade'] = '40-44'
df_netflix.loc[(df_netflix['Age'] >= 45) & (df_netflix['Age'] < 50), 'grupo_idade'] = '45-49'
df_netflix.loc[(df_netflix['Age'] >= 50) & (df_netflix['Age'] < 55), 'grupo_idade'] = '50-54'
df_netflix.loc[(df_netflix['Age'] >= 55) & (df_netflix['Age'] < 60), 'grupo_idade'] = '55-59'
df_netflix.loc[df_netflix['Age'] >= 60, 'grupo_idade'] = '60+'
df_netflix
```

```python
sns.countplot(data=df_netflix[['Subscription Type']].sort_values('Subscription Type'), x='Subscription Type')

 #countplot já traz a contagem, não traz a média como o barplot

plt.show()
```

```python
sns.countplot(data=df_netflix.sort_values('grupo_idade'), x='grupo_idade', hue='Subscription Type') #no hue você passa uma coluna para as cores

plt.show()
```

```python
#vendo o gráfico de outra forma

sns.countplot(data=df_netflix.sort_values('grupo_idade'), x='Subscription Type', hue='grupo_idade') 

plt.show()
```

```python
plt.figure(figsize=(12, 6))  #12 de largura e 6 de comprimento

sns.countplot(data = df_netflix.sort_values('Subscription Type'),
                    x= 'Country',
                    hue='Subscription Type')
                    
plt.title('Países por planos')
plt.xlabel('País')
plt.ylabel('Contagem')

plt.xticks(rotation=330)

plt.legend(title = 'Planos')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  #12 de largura e 6 de comprimento

sns.countplot(data = df_netflix.sort_values('grupo_idade'),
                    x= 'Country',
                    hue='grupo_idade')
                    
plt.title('Países por faixa etária')
plt.xlabel('País')
plt.ylabel('Contagem')

plt.xticks(rotation=330)

plt.legend(title = 'Faixa etária')
plt.show()
```

```python
order_subs_type = ['Basic', 'Standard', 'Premium']

sns.countplot(data=df_netflix.sort_values("grupo_idade"), x="Subscription Type", hue="grupo_idade", palette="viridis", order=order_subs_type)

plt.show()
```
